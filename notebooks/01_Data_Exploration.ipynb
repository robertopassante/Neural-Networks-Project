# 🌍 Satellite Data Exploration: DeepGlobe Dataset

This notebook focuses on the **Data Preparation phase**, similar to the `SYNgen_final.ipynb` but tailored for satellite segmentation. We will explore the classes, extract the dataset, and visualize the ground truth masks.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from data.dataset_manager import extract_dataset, find_train_dir
from data.preprocessing import COLOR_TO_CLASS

# 1. Paths
ZIP_FILE = "DeepGlobe Land Cover Classification Dataset.zip"
DATA_DIR = "extracted_dataset"

# 2. Extraction
if os.path.exists(ZIP_FILE):
    extract_dataset(ZIP_FILE, DATA_DIR)
else:
    print("[!] ZIP file not found. Ensure it is in the root directory.")

train_dir = find_train_dir(DATA_DIR)
print(f"[*] Dataset ready at: {train_dir}")
```

### Visualizing a Sample Pair

```python
if train_dir:
    all_files = [f for f in os.listdir(train_dir) if f.endswith('_sat.jpg')]
    sample_name = all_files[0]
    mask_name = sample_name.replace('_sat.jpg', '_mask.png')

    img = Image.open(os.path.join(train_dir, sample_name))
    mask = Image.open(os.path.join(train_dir, mask_name))

    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    ax[0].imshow(img)
    ax[0].set_title("Satellite Image")
    ax[0].axis("off")
    
    ax[1].imshow(mask)
    ax[1].set_title("Ground Truth Mask")
    ax[1].axis("off")
    
    plt.show()
```
