# 🛰️ SatMAE++ Segmentation Training

In this notebook, we follow the **Segmentation Phase**, using the **SatMAE++ backbone** as required by the **Neural Networks Projects** Notion page. This notebook is inspired by the `Segmentation_final.ipynb` of the reference project.

```python
import torch
import os
from data.dataset_manager import get_dataloader, find_train_dir
from models.satmae_unet import SatMAESegmentor
from training.trainer import SatMAETrainer

# 1. Config
BATCH_SIZE = 32
LEARNING_RATE = 1e-4
NUM_EPOCHS = 15
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
DATA_DIR = "extracted_dataset"

# 2. Prepare DataLoader
train_dir = find_train_dir(DATA_DIR)
if not train_dir:
    raise ValueError("[!] Dataset not found. Please run 01_Data_Exploration.ipynb or extract manually.")

train_loader = get_dataloader(train_dir, batch_size=BATCH_SIZE)
print(f"[*] DataLoader ready: {len(train_loader)} batches.")
```

### Initializing the SatMAE++ Model & Trainer

```python
model = SatMAESegmentor(num_classes=7).to(DEVICE)
trainer = SatMAETrainer(model, train_loader, DEVICE, lr=LEARNING_RATE, checkpoint_dir="checkpoints")

print("[*] Model initialized. Ready for training.")
```

### Executing the Training Phase

```python
# Start training
trainer.train(num_epochs=NUM_EPOCHS)
```
